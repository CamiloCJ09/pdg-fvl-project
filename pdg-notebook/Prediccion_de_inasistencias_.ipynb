## PCA

```python
# Preprocesado y modelado
# ==============================================================================

from sklearn.decomposition import PCA
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import scale
```

```python
#Para realizar el análisis de componentes principales, es recomendable no tener valores nulos
#Verificación de variables con valores nulos

pca_df = feature_df
#Hacemos una copia del Dataframe feature_df, para trabajar en el PCA

pca_df.isna().sum().sort_values()
```

Para tener un primer acercamiento al análisis de componentes principales, imputaremos los valores faltantes con valor de la media en las variables de Sexo, Voluntario, RS/RC y Estrato Socioeconómico. Por el momento, no se usará Grupo étnico para este primer análisis.

```python
#Eliminamos la columna de Grupo Étnico
pca_df = pca_df.drop(["Grupo étnico"],axis=1)
```

```python
#Imputaremos  los valores nulos con la moda de cada columna, ya que tenemos variables categóricas (con dos valores únicos)
pca_df_filled = pca_df.fillna(pca_df.mode().iloc[0])

#Revisamos de nuevo los datos faltantes
pca_df_filled.isna().sum().sort_values()
pca_df_filled.head()
```

### Codificación de variables categóricas para el PCA

```python
#Identificación de valores únicos de las 3 variables categóricas
sex_categories = pca_df_filled['Sexo'].unique()
voluntario_categories = pca_df_filled['Voluntario'].unique()
rs_rc_categories = pca_df_filled['RS / RC'].unique()

print("Valores únicos de Sexo:", sex_categories)
print("Valores únicos de Voluntario:", voluntario_categories)
print("Valores únicos de RS/RC:", rs_rc_categories)
print("Valos únicos de Parte del cuerpo:", pca_df_filled['Parte del Cuerpo'].unique())
```

```python
# Crea un diccionario de mapeo para cada variable categórica
mapeo_sexo = {'Masculino': 0, 'Femenino': 1}
mapeo_voluntario = {'Voluntario': 0, 'Obligatorio': 1}
mapeo_rs_rc = {'RS': 0, 'RC': 1, 'OT': 2}
mapeo_parte_del_cuerpo = {'CEREBRO': 0, 'MMII': 1, 'COLUMNA': 2, 'MMSS': 3, 'OTRO': 4, 'PROSTATA': 5, 'ABDOMEN': 6,
                          'PELVIS': 7, 'MAMA': 8, 'CORAZON': 9}

# Aplica la codificación utilizando el método replace
pca_df_filled['Sexo'] = pca_df_filled['Sexo'].replace(mapeo_sexo)
pca_df_filled['Voluntario'] = pca_df_filled['Voluntario'].replace(mapeo_voluntario)
pca_df_filled['RS / RC'] = pca_df_filled['RS / RC'].replace(mapeo_rs_rc)
pca_df_filled['Parte del Cuerpo'] = pca_df_filled['Parte del Cuerpo'].replace(mapeo_parte_del_cuerpo)

# Muestra las primeras filas del DataFrame resultante
pca_df_filled.head()
```

```python
pca_df_filled.info()
```

### Entrenamiento de modelo PCA y Escalado de variables

#### Con Pipeline

```python
# Entrenamiento modelo PCA con escalado de los datos
# ==============================================================================
pca_pipe = make_pipeline(StandardScaler(), PCA())
pca_pipe
```

```python
pca_pipe.fit(pca_df_filled)
pca_pipe.named_steps
```

```python
# Se extrae el modelo entrenado del pipeline
modelo_pca = pca_pipe.named_steps['pca']
type(modelo_pca)
pca_df_filled.shape
```

```python
# Se convierte el array a dataframe para añadir nombres a los ejes.
pca_components_df = pd.DataFrame(
    data    = modelo_pca.components_,
    columns = pca_df_filled.columns
)
pca_components_df
```

```python
# Heatmap componentes
# ==============================================================================
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(8, 6))
componentes = modelo_pca.components_
plt.imshow(componentes.T, cmap='gray', aspect='auto')
plt.yticks(range(len(pca_df_filled.columns)), pca_df_filled.columns)
plt.xticks(range(len(pca_df_filled.columns)), np.arange(modelo_pca.n_components_) + 1)
plt.grid(False)
plt.colorbar();
```

```python
# Porcentaje de varianza explicada por cada componente
# ==============================================================================
print('----------------------------------------------------')
print('Porcentaje de varianza explicada por cada componente')
print('----------------------------------------------------')
print(modelo_pca.explained_variance_ratio_)

fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(10, 8))
ax.bar(
    x      = np.arange(modelo_pca.n_components_) + 1,
    height = modelo_pca.explained_variance_ratio_
)

for x, y in zip(np.arange(len(pca_df_filled.columns)) + 1, modelo_pca.explained_variance_ratio_):
    label = round(y, 2)
    ax.annotate(
        label,
        (x,y),
        textcoords="offset points",
        xytext=(0,10),
        ha='center'
    )

ax.set_xticks(np.arange(modelo_pca.n_components_) + 1)
ax.set_ylim(0, 1.1)
ax.set_title('Porcentaje de varianza explicada por cada componente')
ax.set_xlabel('Componente principal')
ax.set_ylabel('Por. varianza explicada');
```

```python
var_exp=modelo_pca.explained_variance_ratio_
cum_var_exp = np.cumsum(var_exp) 
var_exp

plt.figure(figsize=(15, 7))
plt.bar(range(len(var_exp)), var_exp, alpha=0.3333, align='center', label='Varianza explicada por cada PC', color = 'g')
plt.step(range(len(cum_var_exp)), cum_var_exp, where='mid',label='Varianza explicada acumulada')
plt.ylabel('Porcentaje de varianza explicada')
plt.xlabel('Componentes principales')
plt.legend(loc='best')
plt.show()
```

## Definición de Funciones

Funciones para vizualizar el rendimiento

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn.metrics import cohen_kappa_score
```

```python
def metrics(target,prediction):
  conf_matrix = confusion_matrix(target,prediction)
  score = accuracy_score(target, prediction)
  report = classification_report(target, prediction,output_dict=True)
  kappa = cohen_kappa_score(target,prediction)

  metrics = [conf_matrix, score, report,kappa]
  return metrics
```

```python
def plot_confusion_matrix(conf_matrix):
    plt.figure(figsize=(8, 6))
    sns.heatmap(conf_matrix, annot=True, cmap="Blues", fmt="d", cbar=False)
    plt.xlabel('Predicted labels')
    plt.ylabel('True labels')
    plt.title(f'Confusion Matrix')
    plt.show()
```

```python
def visualize_report(report):
    # Define custom colors
    colors = ['#ffdfdf', '#dfffff', '#dfffdf', '#dfdfff', '#ffdfff', '#ffffdf']

    # Extracting metrics for each class
    metrics = ['precision', 'recall', 'f1-score', 'support']
    data = {metric: [] for metric in metrics}
    labels = []

    # Include class-specific metrics
    for cls, metrics_values in report.items():
        if cls.isdigit() or cls in ['macro avg', 'weighted avg']:
            labels.append(cls)
            for metric in metrics:
                data[metric].append(metrics_values.get(metric, None))

    # Convert data to DataFrame for easy plotting
    df = pd.DataFrame(data, index=labels)

    # Creating subplots for each metric
    fig, axes = plt.subplots(1, 3, figsize=(18, 6))
    for ax, metric in zip(axes, metrics[:-1]):  # Exclude 'support' for plotting
        sns.barplot(ax=ax, x=df.index, y=metric, data=df, palette=colors)
        ax.set_title(f'{metric.capitalize()} by Class')
        ax.set_ylim(0, 1.1)

    plt.tight_layout()
    plt.show()
```

```python
def plot_kappa(kappa):
  plt.bar(['Cohen\'s Kappa'], [kappa])
  plt.ylabel('Cohen\'s Kappa')
  plt.title('Cohen\'s Kappa Score')
  plt.show()
```

Fine tuning for XGBoost

```python
space={'max_depth': hp.quniform("max_depth", 3, 18, 1),
        'gamma': hp.uniform ('gamma', 1,9),
        'reg_alpha' : hp.quniform('reg_alpha', 40,180,1),
        'reg_lambda' : hp.uniform('reg_lambda', 0,1),
        'colsample_bytree' : hp.uniform('colsample_bytree', 0.5,1),
        'min_child_weight' : hp.quniform('min_child_weight', 0, 10, 1),
        'n_estimators': 180,
        'seed': 0
    }
```

```python
# Define the objective function to minimize
def objective(params):
    params['max_depth'] = int(params['max_depth'])
    xgb_model = xgb.XGBClassifier(**params)
    xgb_model.fit(X_train, y_train)
    y_pred = xgb_model.predict(X_test)
    score = accuracy_score(y_test, y_pred)
    return {'loss': -score, 'status': STATUS_OK}

# Perform the optimization
best_params = fmin(objective, space, algo=tpe.suggest, max_evals=100)
print("Best set of hyperparameters: ", best_params)
```

```python
# instantiate the classifier
best_params['max_depth'] = int(best_params['max_depth'])
xgb_clf = xgb.XGBClassifier(**best_params)
evalset = [(X_train, y_train), (X_test,y_test)]
```

```python
# fit the classifier to the training data
xgb_clf.fit(X_train, y_train, eval_metric='logloss', eval_set=evalset)
```

```python
y_pred = xgb_clf.predict(X_test)
```

```python
results = xgb_clf.evals_result()
```

```python
print('XGBoost model accuracy score: {0:0.4f}'. format(accuracy_score(y_test, y_pred)))
```

OOB Error

```python
oob_error = []
n_estimators_range = range(1, 50)

for n_estimators in n_estimators_range:
    # Crea el modelo RandomForest con el número de estimadores dado
    rf.set_params(n_estimators=n_estimators, oob_score=True, random_state=42)
    rf.fit(X_train, y_train)
    # Obtiene el error OOB y lo agrega a la lista
    oob_error.append(1 - rf.oob_score_)

# Grafica el error OOB en función del número de árboles
plt.figure(figsize=(10, 6))
plt.plot(n_estimators_range, oob_error, label='Out-of-Bag error')
plt.xlabel('Número de árboles')
plt.ylabel('Error OOB')
plt.title('Out-of-Bag Error vs Número de Árboles en RandomForest')
plt.legend()
plt.grid(True)
plt.show()
```
